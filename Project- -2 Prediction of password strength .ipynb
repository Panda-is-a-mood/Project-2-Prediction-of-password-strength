```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib as plt 
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('data.csv',error_bad_lines=False)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(data['strength'])             #this will not work with excel files / graph will be distorted
```

```python
arr = np.array(data)
arr
```

```python
#'password' is independent feature and 'strength' is dependent feature , hence we've passed in onto array to fetch them separately
```

```python
#to provide  robustness to the model , data is shuffled 
```

```python
import random
random.shuffle(arr)
```

```python
#extracting independent and dependent features : 
```

```python
x=[i[0] for i in arr]   #extracting 'password'
y=[i[1] for i in arr]   #extracting 'strength'
```

```python
x
```

```python
#preprocessing 
```

```python
def word(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word('kzde5577')       #separating each alphabet 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer         #applying TF-IDF modelling 
```

```python
vector=TfidfVectorizer(tokenizer=word)
```

```python
X=vector.fit_transform(x)
```

```python
X.shape                     #separated all the alphabets in 'password'
```

```python
vector.get_feature_names()
```

```python
first_doc=X[0]
first_doc
```

```python
first_doc.T.todense()
```

```python
df=pd.DataFrame(first_doc.T.todense(),index=vector.get_feature_names(),columns=['TF-idf'])
df.sort_values(by=['TF-idf'],ascending=False)
```

```python
#Logistic Regression - ML algo
```

```python
from sklearn.model_selection import train_test_split  #splitting data
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
dt=np.array(['345!%adbc'])            #prediction 
pred=vector.transform(dt)
clf.predict(pred)
```

```python
y_pred=clf.predict(X_test)         #prediction using test model 
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
